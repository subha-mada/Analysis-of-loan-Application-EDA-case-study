```python
def negtopos(days):
    if days < 0:
        return(abs(days))
    else:
        return(days)

    

def years(number_of_days): 
    # Assume that years is 
    # of 365 days 
    year = int(number_of_days / 365) 
#     month = int((number_of_days % 365) / Month_day)
#     days = (number_of_days % 365) % DAYS_IN_WEEK 
      
    return(year)
```

```python
app_data['DAYS_BIRTH'] = app_data['DAYS_BIRTH'].apply(negtopos)
app_data['BIRTH_YEAR'] = app_data['DAYS_BIRTH'].apply(years)

app_data['DAYS_EMPLOYED'] = app_data['DAYS_EMPLOYED'].apply(negtopos)
app_data['DAYS_REGISTRATION'] = app_data['DAYS_REGISTRATION'].apply(negtopos)
app_data['DAYS_ID_PUBLISH'] = app_data['DAYS_ID_PUBLISH'].apply(negtopos)
app_data['DAYS_LAST_PHONE_CHANGE'] = app_data['DAYS_LAST_PHONE_CHANGE'].apply(negtopos)

app_data.head()
```

## 4.4-Find correlation for numerical columns of both the cases i.e 0 and 1

```python
# This is for Target_0 df having target as 0

corr = Target_0.corr()
corr = corr.where(np.triu(np.ones(corr.shape), k=1).astype(np.bool))
corrdf0 = corr.unstack().reset_index()
corrdf0.columns = corrdf0.columns = ['VAR1', 'VAR2', 'Correlation']
corrdf0.dropna(subset= ['Correlation'], inplace = True)
corrdf0['Correlation'] = round(corrdf0['Correlation'] , 2)
corrdf0['Correlation'] = corrdf0['Correlation'].abs()
corrdf0.sort_values(by='Correlation', ascending=False).head(15)
```

```python
# This is for Target_1 df having target as 1

corr = Target_1.corr()
corr = corr.where(np.triu(np.ones(corr.shape), k=1).astype(np.bool))
corrdf1 = corr.unstack().reset_index()
corrdf1.columns = corrdf1.columns = ['VAR1', 'VAR2', 'Correlation']
corrdf1.dropna(subset= ['Correlation'], inplace = True)
corrdf1['Correlation'] = round(corrdf1['Correlation'] , 2)
corrdf1['Correlation'] = corrdf1['Correlation'].abs()
corrdf1.sort_values(by='Correlation', ascending=False).head(15)
```
